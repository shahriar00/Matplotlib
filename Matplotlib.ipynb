```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English','Bangla','Hindi','German','Spanish']
people = [111,212,124,156,175]

ax.bar(language,people)
plt.xlabel('Language')
plt.ylabel('People')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English','Bangla','Hindi','German','Spanish']
people = [111,212,124,156,175]

ax.pie(people,labels = language, autopct = "%1.1f%%")
plt.xlabel('Language')
plt.ylabel('People')
```
